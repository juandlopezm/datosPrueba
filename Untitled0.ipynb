**Analisis de pasajeros(Prueba)**

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")

selected_months = flights.query("month in ['Apr', 'Jul']")

plt.figure(figsize=(8, 5))
sns.lineplot(data=selected_months, x="year", y="passengers", hue="month", marker='o')

plt.xlabel("Año")
plt.ylabel("Pasajeros")
plt.title("Número de pasajeros en Abril y Julio a lo largo de los años")
plt.grid(True)

plt.show()
```

**Ventas de productos**

```python
import matplotlib.pyplot as plt
# Paso 2: Datos de ventas por ciudad
productos = ['Pr A', 'Pr B', 'Pr C','Pr D','Pr E','Pr F','Pr G','Pr H',]
Bogota = [150, 100, 90,234,321,455,222,111]
Cali = [200, 180, 150,222,333,444,555,666]
Bucaramanga = [250, 220, 300,400,432,234,123,234]
# Paso 3: Crear gráfico de barras apiladas
plt.bar(productos, Bogota, label='Bogota', color='blue')
plt.bar(productos, Cali, bottom= Bogota, label='Cali', color='green')
plt.bar(productos, Bucaramanga, bottom=[i+j for i,j in zip(Bogota, Cali)], label='Bucaramanga', color='red')

# Paso 4: Añadir etiquetas, título y leyenda
plt.xlabel('Productos')
plt.ylabel('Ventas (Unidades)')
plt.title('Comparación de Ventas por Producto y Ciudad')
plt.legend()
```

**Ventas x ciudades**

```python
import matplotlib.pyplot as plt

productos = ["A", "B", "C", "D", "E", "F", "G", "H"]

bogota = [150, 100,90,120,75,100,90,110]
cali = [200, 180,150,120,75,10,90,33]
bucaramanga = [250, 220,300,60,75,40,90,110]

plt.bar(productos,bogota,color="#f1948a")
plt.bar(productos,cali,color="#7dcea0")
plt.bar(productos,bucaramanga,color="#f4d03f")

plt.title("Ventas de productos en diferentes ciudades", fontsize=10, fontweight="bold",color="#21618c")
plt.xlabel("Productos")
plt.ylabel("Ventas")
plt.legend(title="Ciudades")

plt.show()
```

# **EVOLUCION DE LAS VENTAS POR CIUDAD**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# Paso 2: Crear el DataFrame con las ventas por producto y ciudad
datos = pd.DataFrame({ 'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo','Junio', 'Julio', 'Agosto'],
                      'Papa (Cali)': [100, 150, 200, 250, 300,400,300,150], 'Cebolla(Bogota)': [90, 120, 160, 200, 240,100,200,150],
                       'Tomate(Cartagena)': [80, 110, 150, 190, 230,330,350,240] })
# Paso 3: Configurar el estilo de Seaborn
sns.set(style="whitegrid")
# Crear el gráfico de líneas con Matplotlib
plt.plot(datos['Mes'], datos['Papa (Cali)'], marker='o', color='blue', label='Papa (Cali)')
plt.plot(datos['Mes'], datos['Cebolla(Bogota)'], marker='s', color='green', label='Cebolla(Bogota)')
plt.plot(datos['Mes'], datos['Tomate(Cartagena)'], marker='^', color='red', label='Tomate(Cartagena)')

# Paso 4: Personalizar el gráfico con títulos y etiquetas en español
plt.title('Evolución de ventas por producto y ciudad', fontsize=14)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Ventas (Unidades)', fontsize=12)

# Añadir leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```
